```python
import pandas as pd
from scipy import stats

df = pd.DataFrame(
    data=[
        ['a', 'c'],
        ['a', 'c'],
        ['a', 'c'],
        ['a', 'c'],
        ['b', 'd'],
        ['b', 'd'],
        ['b', 'd'],
        ['b', 'd'],
    ],
    columns=('A', 'B')
)

stats.chi2_contingency(pd.crosstab(df['A'], df['B']))

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

%matplotlib inline

font = {'family' : 'normal',
        'size'   : 19}

plt.rc('font', **font)

fig, ax = plt.subplots(figsize=(16, 11))

x = np.linspace(0, 1, 100)

ax.plot(x, stats.beta.pdf(x, 0.07, 0.07), 'r-', lw=3, alpha=0.8, label='M = 1')
ax.plot(x, stats.beta.pdf(x, 5, 5), 'b-', lw=3, alpha=0.8, label='M = 0')
ax.plot(x, stats.beta.pdf(x, 1, 1), 'g-', lw=3, alpha=0.8, label='M = 0.5')

plt.xticks([0, 0.5, 1], ['0', 'P(X) x P(Y)', 'min(P(X), P(Y))'])

ax.set_xlabel('Selectivity')

ax.legend(loc='best')
```
